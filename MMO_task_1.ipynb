# **IV. Визуализация распределения переменных**

1.   Гистограмма для числовой переменной Attack:




```python
sn.histplot(df['Attack'], kde = True)
plt.title('Распределение Attack')
plt.xlabel('Attack')
plt.ylabel('Частота')
plt.show()
```

2.   Столбчатая диаграмма для категориальной переменной Type 1:




```python
sn.countplot(x = 'Type 1', data = df)
plt.title('Распределение Type 1')
plt.xlabel('Type 1')
plt.ylabel('Количество')
plt.show()
```

3.   Боксплот для числовой переменной Total:




```python
sn.boxplot(x = df['Total'])
plt.title('Боксплот Total')
plt.xlabel('Total')
plt.show()
```

**Мои наблюдения:**
1. Гистограмма распределения по атаке практически соответствует нормальному распределению, потому что среднее значение расположено между 75 и 100.
2. Столбчатая диаграмма по типам первой категории показывает, что в среднем больше всего покемонов относится к водной стихии, наименьшее количество - к летающим.
3. На боксплоте по суммарному количеству очков характеристик видно, что отсутствуют выбросы. Медиана расположена между 400 и 500 значением. Скорее всего отсутствие выбросов связано с тем, что у покемонов ни одна из характеристик не может быть равна 0.

P.s. По факту, для столбчатой диаграммы можно было взять ещё Generation. Это числовая переменная (потому что её значения составляют целые числа от 1 до 6), но с другой стороны, т. к. эта переменная обозначает категории покемонов, её можно рассмотреть и как категориальную (в другом контексте).
